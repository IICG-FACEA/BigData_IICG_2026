```python
# PRUEBA DE SALUD DEL ENTORNO
import platform
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager

print(f"1. Sistema: {platform.system()} | Arquitectura: {platform.machine()}")
options = Options()
options.add_argument("--headless")
options.add_argument("--no-sandbox")
options.binary_location = "/usr/bin/google-chrome"

try:
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().
    install()), options=options)
    print("2. Driver y Chrome sincronizados.")
    driver.get("https://www.google.com")
    print(f"3. Conexi n a Internet: {driver.title}")
    driver.quit()
    print("4. Entorno listo para trabajar.")
except Exception as e:
    print(f" ERROR EN EL RESET: {e}")
```
