```python
import requests
from bs4 import BeautifulSoup

# Configuracion de entrada
url = "https://books.toscrape.com/"
headers = {"User-Agent": "Mozilla/5.0"} # Para que no nos bloqueen

# 1. Peticion al servidor
respuesta = requests.get(url, headers=headers)

# 2. Convertir el HTML en un objeto de Python
sopa = BeautifulSoup(respuesta.text, "html.parser")

# 3. Buscar los datos 
# Cambia ’h2’ y ’nombre-de-la-clase’ por lo que se inspecciono
resultados = sopa.find_all("article", class_="product_pod")
```

```python
# Lista para guardar todos los datos
todos_los_datos = []

# 4. Recorrer cada libro y extraer la información
for item in resultados:
    # Extraer el título: está dentro de <h3> y luego <a> con atributo 'title'
    titulo_elemento = item.find("h3").find("a")
    titulo = titulo_elemento.get("title", "Sin título")
    
    # Extraer el precio: está dentro de <p class="price_color">
    precio_elemento = item.find("p", class_="price_color")
    precio = precio_elemento.text.strip() if precio_elemento else "Sin precio"
    
    # Crear el diccionario para este libro
    datos_item = {
        "titulo": titulo,
        "valor": precio,
        "grupo": "Retail",
        "fuente": "books.toscrape.com"
    }
    
    # Agregar a la lista general
    todos_los_datos.append(datos_item)

# Mostrar resultados
print(f"He guardado {len(todos_los_datos)} diccionarios.")
print("\n" + "="*50)
print("Primeros 5 libros capturados:")
print("="*50)

for i, libro in enumerate(todos_los_datos[:5], 1):
    print(f"\n Libro {i}:")
    print(f"   Título: {libro['titulo']}")
    print(f"   Precio: {libro['valor']}")
    print(f"   Grupo: {libro['grupo']}")
```
