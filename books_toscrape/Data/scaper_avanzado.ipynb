```python
import requests #librería para peticiones a internet
from bs4 import BeautifulSoup #para buscar dentro del html

url = "http://books.toscrape.com/"
headers = {"User-Agent": "Mozilla/5.0"} #careta

respuesta = requests.get(url, headers=headers) #descarga el html
respuesta.encoding = 'utf-8' #para que se vea bien el simbolo
sopa = BeautifulSoup(respuesta.text, 'html.parser') # html a objt python

resultados = sopa.find_all('article', class_='product_pod')
print(f"Conexión exitosa. Encontré {len(resultados)} libros.")
```

```python
todos_los_datos = []

for item in resultados:
    datos_item = {
        "titulo": item.find('h3').find('a')['title'],
        "precio": item.find('p', class_='price_color').text.strip(), #strip es para eliminar espacios sobrantes
        "disponibilidad": item.find('p', class_='instock').text.strip(),
        "grupo": "E-commerce"
    }
    todos_los_datos.append(datos_item)

print(f"He guardado {len(todos_los_datos)} diccionarios.")
```

```python
for libro in todos_los_datos[:3]:
    print(libro)
```
