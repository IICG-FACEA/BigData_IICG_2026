```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("TransformacionesMasivas") \
    .config("spark.mongodb.read.connection.uri", "mongodb+srv://rodrigoortega_db_user:ucn12@cluster0.uia9vsi.mongodb.net/BigData_UCN.scrapers?retryWrites=true&w=majority") \
    .config("spark.mongodb.write.connection.uri", "mongodb+srv://rodrigoortega_db_user:ucn12@cluster0.uia9vsi.mongodb.net/BigData_UCN.transformados?retryWrites=true&w=majority") \
    .config("spark.jars.packages", "org.mongodb.spark:mongo-spark-connector_2.12:10.2.0") \
    .getOrCreate()
```

```python
df = spark.read.format("mongodb") \
    .option("database", "BigData_UCN") \
    .option("collection", "scrapers") \
    .load()

df.show(5)
```

```python
df_clean = df.dropna(subset=["campo_clave"]) \
             .withColumnRenamed("nombreOriginal", "nombre") \
             .filter(df["edad"] > 18)
df_clean.show(5)
```

```python
df_clean.write.format("mongodb").mode("append").save()
```
