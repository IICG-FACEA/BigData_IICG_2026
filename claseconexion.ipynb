```python
import requests 
import pandas as pd 

# Intentemos capturar datos reales de la API de intercambio de monnedas
try: 
    url = "https://api.exchangerate-api.com/v4/latest/USD"
    respuesta = requests.get(url)
    datos = respuesta.json()
    df = pd.DataFrame(datos['rates'].items(), columns=['Moneda', 'Valor'])
    print(" conexion exitosa! aqui los primeros datos capturados:")
    print(df.head())

except Exception as e:
    print(f"error en la conexion: {e}")
    
```
