```python
import requests
from bs4 import BeautifulSoup
import re

url = "https://www.chileautos.cl/"
headers = {
    "User-Agent": "Mozilla/5.0"
}

respuesta = requests.get(url, headers=headers)
print("Status code:", respuesta.status_code)

sopa = BeautifulSoup(respuesta.text, "html.parser")

texto_completo = sopa.get_text("\n", strip=True)

# Buscar precios chilenos tipo $13,290,000
precios = re.findall(r"\$\d{1,3}(?:,\d{3})+", texto_completo)

print("Primeros precios encontrados:")
for p in precios[:10]:
    print(p)
```
