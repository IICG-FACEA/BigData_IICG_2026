```python
import requests
from bs4 import BeautifulSoup
import re

url = "https://www.chileautos.cl/"
headers = {"User-Agent": "Mozilla/5.0"}

respuesta = requests.get(url, headers=headers)
print("Status code:", respuesta.status_code)

sopa = BeautifulSoup(respuesta.text, "html.parser")

autos = sopa.find_all("a")

resultados = []

for auto in autos:
    texto = auto.get_text(" ", strip=True)

    # Buscar texto que tenga precio y algo más (posible auto)
    if "$" in texto and len(texto) < 120:
        texto = re.sub(r"\s+", " ", texto)
        resultados.append(texto)

# Limpiar duplicados
vistos = set()

print("Autos encontrados:\n")

for r in resultados:
    if r not in vistos:
        print(r)
        vistos.add(r)
    if len(vistos) >= 15:
        break
```

```python
import pandas as pd

datos = []

for r in resultados:
    partes = r.split("desde")
    
    if len(partes) == 2:
        nombre = partes[0].strip()
        precio = partes[1].strip()
        
        datos.append({
            "Auto": nombre,
            "Precio": precio
        })

df = pd.DataFrame(datos)

df.to_csv("dataset.csv", index=False)

print("Archivo guardado ✔")
df.head()
```
