```python
import requests
from bs4 import BeautifulSoup

# Configuraci n de entrada
url = "https://books.toscrape.com/"
headers = {"User-Agent": "Mozilla/5.0"} # Para que no nos bloqueen

# 1. Peticion al servidor
respuesta = requests.get(url, headers=headers)

# 2. Convertir el HTML en un objeto de Python
sopa = BeautifulSoup(respuesta.text, "html.parser")

# 3. Buscar los datos (Ejemplo: buscar todos los t tulos)
# Cambia ’h2’ y ’nombre-de-la-clase’ por lo que inspeccionaste
resultados = sopa.find_all("article", class_="product_pod")

# 4. Mostrar en pantalla
for dato in resultados:
  print(f"Encontrado: {dato.text.strip()}")
```
