```python
import requests
from bs4 import BeautifulSoup

# Configuraci n de entrada
url = "https://books.toscrape.com/"
headers = {"User-Agent": "Mozilla/5.0"} # Para que no nos bloqueen

# 1. Peticion al servidor
respuesta = requests.get(url, headers=headers)

# 2. Convertir el HTML en un objeto de Python
sopa = BeautifulSoup(respuesta.text, "html.parser")

# 3. Buscar los datos (Ejemplo: buscar todos los t tulos)
# Cambia ’h2’ y ’nombre-de-la-clase’ por lo que inspeccionaste
resultados = sopa.find_all("div", class_="product_price")

# 4. Mostrar en pantalla
for dato in resultados:
  print(f"Encontrado: {dato.text.strip()}")
```

```python
todos_los_datos = [] # Lista vacia para guardar todo

for item in resultados:
   # Creamos el diccionario para ESTE producto
   datos_item = {
        "titulo": item.find("div").text.strip(),
        "valor": item.find(", class_="price_color").text.strip(),
        "grupo": "Retail" # Etiqueta de tu caso de estudio
}

 # Lo agregamos a nuestra lista general
 todos_los_datos.append(datos_item)

# Ahora tenemos una "colecci n" lista para MongoDB
print(f"He guardado {len(todos_los_datos)} diccionarios.")
```
