```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager

# ==========================================================
# 1. MOTOR DE NAVEGACIÓN
# ==========================================================
def iniciar_navegador():
    options = Options()
    options.binary_location = "/usr/bin/google-chrome"
    options.add_argument("--headless")
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")
    
    return webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

# ==========================================================
# 2. CONFIGURACIÓN
# ==========================================================
URL_OBJETIVO = "https://books.toscrape.com/" 

SELECTOR_CONTENEDOR = "article.product_pod" 
SELECTOR_DATO_A     = "h3 a"
SELECTOR_DATO_B     = "p.price_color"

# ==========================================================
# 3. SCRAPING CON PAGINACIÓN
# ==========================================================
driver = iniciar_navegador()
dataset_final = []

try:
    print("Conectando a la fuente de datos...")
    driver.get(URL_OBJETIVO)
    time.sleep(3) #tiempo de espera

    pagina = 1
    while True:
        print(f"Procesando página {pagina}...")

        elementos = driver.find_elements(By.CSS_SELECTOR, SELECTOR_CONTENEDOR) #bloques de informacion

        for item in elementos:
            try:
                columna_a = item.find_element(By.CSS_SELECTOR, SELECTOR_DATO_A).text
                columna_b = item.find_element(By.CSS_SELECTOR, SELECTOR_DATO_B).text

                valor_limpio = "".join(filter(str.isdigit, columna_b)) #limpieza basica de caracteres no numericos

                dataset_final.append({
                    "variable_nombre": columna_a,
                    "variable_valor": valor_limpio,
                    "status": 1.0
                })
            except:
                continue #si un elemento esta vacio o es anuncion, saltar al siguiente

        # BOTÓN SIGUIENTE
        try:
            boton_siguiente = driver.find_element(By.XPATH, '//li[@class="next"]/a')
            boton_siguiente.click()
            time.sleep(2)
            pagina +=1 #suma 1 a la pagina al procesar
        except:
            print("Última página alcanzada")
            break

    # ==========================================================
    # 4. GUARDAR RESULTADOS
    # ==========================================================
    df = pd.DataFrame(dataset_final)
    nombre_archivo = "datos_books.csv"
    df.to_csv(nombre_archivo, index=False)

    print(f"\nProceso exitoso")
    print(f"Total de datos: {len(df)}")
    print(df.head())

finally:
    driver.quit()
    print("Navegador cerrado")
```
