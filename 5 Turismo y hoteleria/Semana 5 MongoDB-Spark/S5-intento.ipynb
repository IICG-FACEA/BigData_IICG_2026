```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F

# 1. Crear sesión de Spark con conexión a MongoDB
spark = SparkSession.builder \
    .appName("Analisis_Categorias") \
    .config("spark.mongodb.read.connection.uri", "mongodb://bigdata_mongodb:27017/proyecto_bigdata.datos_scraping") \
    .config("spark.mongodb.write.connection.uri", "mongodb://bigdata_mongodb:27017/proyecto_bigdata.datos_scraping") \
    .config("spark.jars.packages", "org.mongodb.spark:mongo-spark-connector_2.12:10.6.1") \
    .getOrCreate()

# 2. Cargar datos desde la colección datos_scraping
df = spark.read.format("mongodb").load()

# 3. Mostrar esquema y primeras filas
df.printSchema()
df.show(5)
```

```python
# Contar registros
print("Total registros:", df.count())

# Agrupar por grupo y calcular promedio de valor
df.groupBy("grupo").agg(F.avg("valor").alias("promedio_valor")).show()

# Filtrar libros con precio mayor a 50
df.filter(df["valor"] > 50).show(10)
```

```python
# Definir la ruta de salida dentro de tu carpeta del proyecto
output_path = "/home/jovyan/5 Turismo y hoteleria/datos_scraping_export.csv"

# Exportar el DataFrame a CSV con encabezados
df.write.option("header", True).mode("overwrite").csv(output_path)

print("Archivo CSV exportado en:", output_path)
```
