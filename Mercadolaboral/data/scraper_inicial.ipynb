```python
import requests
from bs4 import BeautifulSoup

# Configuración de entrada
url = "https://remoteok.com/remote-jobs/remote-senior-frontend-engineer-level-1130614"
headers = {"User-Agent": "Mozilla/5.0"}  # Para que no nos bloqueen

# 1. Petición al servidor
respuesta = requests.get(url, headers=headers)

# 2. Convertir el HTML en un objeto de Python
sopa = BeautifulSoup(respuesta.text, 'html.parser')

# 3. Buscar los datos (ejemplo: ubicaciones de ofertas de trabajo)
# Cambia 'div' y 'location' por lo que inspeccionaste
resultados = sopa.find_all("div", class_="salary")

# 4. Mostrar en pantalla
for dato in resultados[:10]:  # mostramos los primeros 10
    print(f"Encontrado: {dato.text.strip()}")
```
