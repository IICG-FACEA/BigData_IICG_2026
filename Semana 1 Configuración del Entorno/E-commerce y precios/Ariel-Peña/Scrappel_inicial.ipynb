```python
import requests
from bs4 import BeautifulSoup

# Configuracion de entrada
url = "http://books.toscrape.com/"
headers = {"User-Agent": "Mozilla/5.0"} #para que no nos bloqueen

# 1. Peticion al servidor
respuesta = requests.get(url, headers=headers)

# 2. Convertir el HTML en un objeto de Python
sopa = BeautifulSoup(respuesta.text, 'html.parser')

# 3. Buscar los datos
productos = sopa.find_all("article", class_="product_pod")

# 4. Recorrer cada producto y extraer datos
for producto in productos: 
    titulo = producto.h3.a["title"]
    precio = producto.find("p", class_="price_color").get_text()
    stock = producto.find("p", class_="instock availability").get_text(strip=True)
    rating = producto.find("p", class_="star-rating")["class"][1]

    print(f"Título: {titulo}, Precio: {precio}, Stock: {stock}, Rating: {rating}")
```
