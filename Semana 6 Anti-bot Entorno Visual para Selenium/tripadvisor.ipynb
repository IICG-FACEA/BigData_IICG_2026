```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import time

# Configura el navegador para usarlo dentro del contenedor
options = Options()
options.binary_location = "/usr/bin/google-chrome"
options.add_argument("--no-sandbox")                 # Evita errores comunes en Docker
options.add_argument("--disable-dev-shm-usage")     # Reduce fallos por memoria compartida
options.add_argument(
    "user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
)                                                   # Simula un navegador común

# Inicia el navegador automatizado
driver = webdriver.Chrome(options=options)

# URL pública del caso Tripadvisor
url = "https://www.tripadvisor.cl/Hotels-g294305-Santiago_Santiago_Metropolitan_Region-Hotels.html"
driver.get(url)

time.sleep(10)                                      # Espera para que cargue la página

# Diagnóstico básico
print("Título:", driver.title)                      # Muestra el título de la página
print("URL actual:", driver.current_url)            # Muestra la URL final
print("¿Aparece captcha?:", "captcha" in driver.page_source.lower())  # Busca señales simples de bloqueo

# Prueba distintos selectores generales para explorar la página
selectores = [
    "a",
    "div",
    "article",
    '[class*="hotel"]',
    '[class*="result"]'
]

for s in selectores:
    elementos = driver.find_elements(By.CSS_SELECTOR, s)
    print(f"{s} -> {len(elementos)} elementos")     # Cuenta cuántos elementos devuelve cada selector

driver.quit()                                       # Cierra el navegador
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import time

options = Options()
options.binary_location = "/usr/bin/google-chrome"

options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")
options.add_argument("--disable-gpu")
options.add_argument("--disable-software-rasterizer")
options.add_argument("--window-size=1920,1080")

options.add_argument(
    "user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
)

driver = webdriver.Chrome(options=options)

url = "https://www.tripadvisor.cl/Hotels-g294305-Santiago_Santiago_Metropolitan_Region-Hotels.html"
driver.get(url)

time.sleep(10)

html = driver.page_source.lower()

print("Título:", driver.title)
print("URL actual:", driver.current_url)

#  detección de bloqueo
if "access is temporarily restricted" in html:
    print(" Bloqueo detectado: acceso restringido por el sitio")
elif "captcha" in html:
    print("⚠ CAPTCHA detectado")
else:
    print(" Página cargada sin bloqueo evidente")

time.sleep(20)
driver.quit()
```
