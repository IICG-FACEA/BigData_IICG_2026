```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import time

options = Options()
options.binary_location = "/usr/bin/google-chrome"

options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")
options.add_argument(
    "user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
)

driver = webdriver.Chrome(options=options)

url = "https://www.booking.com/searchresults.es.html?ss=Santiago"
driver.get(url)

time.sleep(10)

print("Título:", driver.title)
print("URL actual:", driver.current_url)
print("¿Aparece captcha?:", "captcha" in driver.page_source.lower())

input("Observa el entorno visual y presiona Enter...")
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import time

# Configura las opciones del navegador
options = Options()
options.binary_location = "/usr/bin/google-chrome"

options.add_argument("--no-sandbox")                 # Evita errores comunes dentro de Docker
options.add_argument("--disable-dev-shm-usage")     # Reduce fallos por memoria compartida
options.add_argument(
    "user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
)                                                   # Simula un navegador común

# Crea el navegador controlado por Selenium
driver = webdriver.Chrome(options=options)

# URL de búsqueda en Booking
url = "https://www.booking.com/searchresults.es.html?ss=Santiago"
driver.get(url)                                     # Abre la página

time.sleep(10)                                      # Espera para que cargue el contenido

# Muestra información básica para diagnóstico
print("Título:", driver.title)                      # Título de la página
print("URL actual:", driver.current_url)            # URL final luego de posibles redirecciones
print("¿Aparece captcha?:", "captcha" in driver.page_source.lower())  # Busca la palabra captcha en el HTML

input("Observa el entorno visual y presiona Enter para cerrar...")
driver.quit()                                       # Cierra el navegador
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import time

# Configura Chrome para usarlo dentro del contenedor Docker
options = Options()
options.binary_location = "/usr/bin/google-chrome"
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")
options.add_argument(
    "user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64)"
)

# Inicia el navegador
driver = webdriver.Chrome(options=options)

# Abre la búsqueda de hoteles en Santiago
url = "https://www.booking.com/searchresults.es.html?ss=Santiago"
driver.get(url)

time.sleep(10)  # Espera que la página termine de cargar

# Cuenta elementos div para ver si la página cargó contenido
elementos = driver.find_elements(By.TAG_NAME, "div")
print(f"Elementos encontrados: {len(elementos)}")

# Muestra datos básicos para entender el estado de la página
print("Título:", driver.title)
print("URL actual:", driver.current_url)
print("¿Aparece captcha?:", "captcha" in driver.page_source.lower())

driver.quit()  # Cierra el navegador
```
