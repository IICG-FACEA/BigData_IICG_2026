```python
import time
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By

def iniciar_navegador():
    options = Options()
    options.binary_location = "/usr/bin/google-chrome"
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")
    options.add_argument(
        "user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 Chrome/135.0.0.0 Safari/537.36"
    )
    return webdriver.Chrome(options=options)

driver = iniciar_navegador()

url = "https://www.falabella.com/falabella-cl/category/cat40052/Computadores"
driver.get(url)
time.sleep(8)

productos = driver.find_elements(By.CSS_SELECTOR, '[data-pod="catalyst-pod"]')
print(f"Productos encontrados: {len(productos)}")

for i, producto in enumerate(productos[:5], start=1):
    print(f"\nProducto {i}")
    print(producto.text)

driver.quit()
```

```python
import time
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By

options = Options()
options.binary_location = "/usr/bin/google-chrome"
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")
options.add_argument(
    "user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 Chrome/135.0.0.0 Safari/537.36"
)

driver = webdriver.Chrome(options=options)

url = "https://www.falabella.com/falabella-cl/category/cat40052/Computadores"
driver.get(url)
time.sleep(8)

print("Título:", driver.title)
print("URL actual:", driver.current_url)
print("¿Aparece captcha?:", "captcha" in driver.page_source.lower())

selectores = [
    '[data-pod="catalyst-pod"]',
    '[data-testid="pod"]',
    '.pod',
    '.pod-item',
    'article'
]

for s in selectores:
    elementos = driver.find_elements(By.CSS_SELECTOR, s)
    print(f"{s} -> {len(elementos)} elementos")

driver.quit()
```
