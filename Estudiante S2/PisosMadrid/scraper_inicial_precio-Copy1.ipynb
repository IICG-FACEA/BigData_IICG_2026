```python
import requests
from bs4 import BeautifulSoup

# Configuración de entrada
url = "https://www.pisos.com/venta/pisos-madrid/"
headers = {"User-Agent": "Mozilla/5.0"} # Para que no nos bloqueen, Le dice al servidor qué tipo de cliente está haciendo la petición, para este caso un navegador
                                        # "Referer": "https://www.google.com/": Le indica al servidor desde qué página aparentemente viene el usuario. 
# 1. Petición al servidor
respuesta = requests.get(url, headers=headers)

# 2. Convertir el HTML en un objeto de Python
sopa = BeautifulSoup(respuesta.text, 'html.parser')

# 3. Buscar los datos (Ejemplo: buscar todos los títulos)
# Cambia 'h2' y 'nombre-de-la-clase' por lo que inspeccionaste
resultados = sopa.find_all('span', class_='ad-preview__price')

# 4. Mostrar en pantalla
for dato in resultados:
    print(f"Encontrado: {dato.text.strip()}")
```
