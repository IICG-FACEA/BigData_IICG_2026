```python
import requests
from bs4 import BeautifulSoup

url = "https://www.pisos.com/venta/pisos-madrid/"
headers = {"User-Agent": "Mozilla/5.0"}

respuesta = requests.get(url, headers=headers)
sopa = BeautifulSoup(respuesta.text, 'html.parser')

#  Contenedor principal
propiedades = sopa.find_all('div', class_='ad-preview')

lista_propiedades = []

for propiedad in propiedades:
    
    #  Precio
    precio_tag = propiedad.find('span', class_='ad-preview__price')
    precio = precio_tag.text.strip() 
    
    #  Teléfono
    telefono_tag = propiedad.find('button', class_='contact-box__phone')
    telefono = telefono_tag.get('data-number') 
    
    #  Diccionario
    propiedad_dict = {
        "precio": precio,
        "telefono": telefono
    }
    
    lista_propiedades.append(propiedad_dict)
    
print(f"Se han guardado {len(lista_propiedades) } diccionarios")

# Mostrar resultados
for p in lista_propiedades:
    print(p)

```
