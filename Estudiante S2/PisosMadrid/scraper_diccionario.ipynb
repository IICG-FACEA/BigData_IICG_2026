```python
import requests
from bs4 import BeautifulSoup

# Configuración de entrada
url = "https://www.pisos.com/venta/pisos-madrid/"
headers = {"User-Agent": "Mozilla/5.0"} # Para que no nos bloqueen, Le dice al servidor qué tipo de cliente está haciendo la petición, para este caso un navegador
                                        # "Referer": "https://www.google.com/": Le indica al servidor desde qué página aparentemente viene el usuario. 
# 1. Petición al servidor
respuesta = requests.get(url, headers=headers)

# 2. Convertir el HTML en un objeto de Python
sopa = BeautifulSoup(respuesta.text, 'html.parser')

# 3. Buscar los datos (Ejemplo: buscar todos los títulos)
# Cambia 'h2' y 'nombre-de-la-clase' por lo que inspeccionaste
prodiedades = sopa.find_all('div', class_='ad-preview')

# 4. Lista donde guardaremos los datos
lista_propiedades = []

# 5. Recorrer cad propiedad
for propiedad in prodiedades:
    # Precio
    precio = propiedad.find('span', class_='ad-preview__price')
    precio = precio.text.strip() if precio else "No disponible"
     # Características (donde vienen m2, habitaciones, baños)
    caracteristicas = propiedad.find_all('p', class_='ad-preview__char')
 
    for c in caracteristicas:
        texto = c.text.lower()        
        if "habs." in texto:
            dormitorios = texto
        elif "m²" in texto:
            metros = texto
        elif "baños" in texto:
            banos = texto
    
    # Crear diccionario
    propiedad_dict = {
        "precio": precio,
        "habitaciones": dormitorios,
        "metros": metros,
        "banos": banos
    }
    
    # Guardar en lista
    lista_propiedades.append(propiedad_dict)

# 6. Mostrar resultados
for p in lista_propiedades:
    print(p)

    

    
    
```
