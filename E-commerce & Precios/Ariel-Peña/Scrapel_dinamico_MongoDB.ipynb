```python
from pymongo import MongoClient

# Conexión desde el contenedor Jupyter al servicio MongoDB
client = MongoClient("mongodb://bigdata_mongodb:27017/")

db = client["proyecto_bigdata"]
coleccion = db["G_Ecommerce_scraping"]

print("Conexión establecida con MongoDB desde Jupyter.")
```

```python
import pandas as pd

# Cargar el archivo CSV (puede ser el de 20 o el de 100 registros)
df = pd.read_csv("E-commerce & Precios/Ariel-Peña/datos_extraidos_grupo.csv")

print("CSV cargado con éxito. Registros:", len(df))
df.head()
```

```python
data_dict = df.to_dict("records")
resultado = coleccion.insert_many(data_dict)

print("Datos insertados en MongoDB con éxito.")
print("Total de documentos en la colección:", coleccion.count_documents({}))

for doc in coleccion.find().limit(5):
    print(doc)

print("Primeros IDs insertados:", resultado.inserted_ids[:5])
```
