```python
import requests
from bs4 import BeautifulSoup

url = "http://books.toscrape.com/"
headers = {"User-Agent": "Mozilla/5.0"}

respuesta = requests.get(url, headers=headers)
sopa = BeautifulSoup(respuesta.content, 'html.parser')

libros = sopa.find_all('article', class_='product_pod')

for libro in libros:
    # Estrellas
    star_tag = libro.find('p', class_='star-rating')
    estrellas = "N/A"
    if star_tag:
        for clase in star_tag['class']:
            if clase != "star-rating":
                estrellas = clase  # One, Two, Three, Four, Five
    
    # Nombre
    nombre_tag = libro.find('h3').find('a')
    nombre = nombre_tag['title'] if nombre_tag and 'title' in nombre_tag.attrs else "N/A"
    
    # Precio
    precio_tag = libro.find('p', class_='price_color')
    precio = precio_tag.text.strip() if precio_tag else "N/A"
    
    # Stock / disponibilidad
    stock_tag = libro.find('p', class_='instock availability')
    stock = stock_tag.text.strip() if stock_tag else "N/A"
    
    print(f"{estrellas} | {nombre} | {precio} | {stock}")
```
