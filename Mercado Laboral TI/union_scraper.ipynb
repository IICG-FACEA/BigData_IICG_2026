```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager

# ==========================================================
# 1. MOTOR DE NAVEGACIÓN (Configuración de Laboratorio)
# ==========================================================
def iniciar_navegador():
    options = Options()
    options.add_argument("--headless")
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")
    options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36")
    
    return webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

# ==========================================================
# 2. CONFIGURACIÓN DEL GRUPO
# ==========================================================
URL_OBJETIVO = "https://books.toscrape.com/"

SELECTOR_CONTENEDOR = "article.product_pod"
SELECTOR_DATO_A     = "h3 a"
SELECTOR_DATO_B     = "p.price_color"
XPATH_SIGUIENTE     = "//li[@class='next']/a"

# ==========================================================
# 3. EJECUCIÓN DEL SCRAPING (Extracción Total)
# ==========================================================
driver = iniciar_navegador()
dataset_final = []
contador_paginas = 1

try:
    print(f" Conectando a la fuente de datos...")
    driver.get(URL_OBJETIVO)

    while True:
        print(f" Procesando Página {contador_paginas}...")
        
        # Espera a que los libros carguen en la página actual
        WebDriverWait(driver, 10).until(
            EC.presence_of_all_elements_located((By.CSS_SELECTOR, SELECTOR_CONTENEDOR))
        )

        elementos = driver.find_elements(By.CSS_SELECTOR, SELECTOR_CONTENEDOR)

        for item in elementos:
            try:
                # Extracción de datos
                columna_a = item.find_element(By.CSS_SELECTOR, SELECTOR_DATO_A).get_attribute("title")
                columna_b = item.find_element(By.CSS_SELECTOR, SELECTOR_DATO_B).text
                valor_limpio = "".join(filter(str.isdigit, columna_b))

                dataset_final.append({
                    "variable_nombre": columna_a,
                    "variable_valor": valor_limpio,
                    "status": 1.0 
                })
            except:
                continue

        # Lógica de navegación infinita
        try:
            # Buscamos el botón "Next"
            boton_siguiente = driver.find_element(By.XPATH, XPATH_SIGUIENTE)
            
            # Scroll al botón y clic
            driver.execute_script("arguments[0].scrollIntoView();", boton_siguiente)
            boton_siguiente.click()
            
            # Espera para que la nueva página cargue (importante)
            time.sleep(2)
            contador_paginas += 1
            
        except:
            # Si no encuentra el botón "Next", el ciclo se rompe
            print(" No se encontró el botón 'Next'. Se ha llegado al final del catálogo.")
            break

    # ==========================================================
    # 4. SALIDA DE DATOS
    # ==========================================================
    if dataset_final:
        df = pd.DataFrame(dataset_final)
        nombre_archivo = "datos_extraidos_completos.csv"
        df.to_csv(nombre_archivo, index=False)
        print(f" Proceso exitoso. Archivo generado: {nombre_archivo}")
        print(f" Total de libros extraídos: {len(df)}")
    else:
        print(" No se capturaron datos.")

finally:
    driver.quit()
    print(" Navegador cerrado.")
```
