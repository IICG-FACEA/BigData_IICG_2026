```python
import requests
from bs4 import BeautifulSoup

url = "https://quotes.toscrape.com/"
headers = {"User-Agent": "Mozilla/5.0"}

respuesta = requests.get(url, headers=headers)
sopa = BeautifulSoup(respuesta.text, 'html.parser')

quotes = sopa.find_all('span', class_='text')

datos = []

for q in quotes:
    datos.append({
        "frase": q.text.strip()
    })

print(datos[:5])
```

```python
quotes = sopa.find_all('div', class_='quote')

datos = []

for q in quotes:
    frase = q.find('span', class_='text').text.strip()
    autor = q.find('small', class_='author').text.strip()

    datos.append({
        "frase": frase,
        "autor": autor
    })

print(datos[:5])
```
